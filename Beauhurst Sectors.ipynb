```python
import glob
import os

path = 'Martha Data Exports' # use your path
all_files = glob.glob(os.path.join(path , "*.csv"))

print(all_files)

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0, low_memory=False)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
# Sectors: Advanced Manufacturing, Creative Industries, Clean Energy, Life Sciences, Defence, Digital and Tech

frame.loc[:, frame.columns.str.contains("Advanced Manufacturing", case=False)]

search_terms = ["Advanced manufacturing", 'Creative industries', 'Clean energy', 'Life sciences', 'Defence', 'Digital and tech']

# Find columns that contain any of the search terms
matching_columns = [col for col in df.columns if any(term in col for term in search_terms)]

print(matching_columns)

sectors_columns = ['Buzzwords - Advanced manufacturing', 'Buzzwords - Clean energy', 'Buzzwords - Creative industries', 'Buzzwords - Digital and technologies', 'Buzzwords - Life sciences', 'Buzzwords - Defence']

data_w_sectors = frame.loc[:, sectors_columns+['Beauhurst URL']]

# Clean column names
data_w_sectors.columns = data_w_sectors.columns.str.replace("Buzzwords - ", "")

data_w_sectors_melt = data_w_sectors.melt(id_vars='Beauhurst URL', var_name='Sectors')

data_w_sectors = data_w_sectors_melt.loc[data_w_sectors_melt['value']].drop(columns='value')

data_w_sectors
```

```python
# Number of companies

data_num = data_w_sectors.groupby('Sectors', as_index=False).nunique()

plt.figure(figsize=(10,5))

sns.barplot(data=data_num.sort_values(by='Beauhurst URL', ascending=False), x='Beauhurst URL', y='Sectors', width=0.6)

ax = plt.gca()
# Add value labels to the end of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3, weight='bold', fontsize=12, fmt="{:,.0f}")

plt.xlabel('Number of Companies')

plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))


plt.tight_layout()
plt.show()

```

```python
# Tracking
# 455 companies Advanced Manufacturing and Comapnies House == Active

df_tracking = frame[['Beauhurst URL', 'Tracked status', 'Tracking reasons - Has received equity funding', 'Tracking reasons - Has completed an MBO/MBI', 
                'Tracking reasons - Has received a qualifying loan', 'Tracking reasons - Has attended an accelerator', 
                'Tracking reasons - Is a 10% scaleup', 'Tracking reasons - Is a 20% scaleup', 
                'Tracking reasons - Was spun out of an academic institution', 'Tracking reasons - Has been featured in a high growth list cohort',
                'Tracking reasons - Has received a large innovation grant']]
```

```python
df_tracked = df_tracking.groupby('Tracked status').agg({'Beauhurst URL':'count'})
```

```python
plt.figure(figsize=(10,5))

sns.barplot(data=df_tracked.reset_index(), x='Beauhurst URL', y='Tracked status', width=0.6)

ax = plt.gca()
# Add value labels to the end of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge', padding=3, weight='bold', fontsize=12, fmt="{:,.0f}")

plt.xlabel('Number of Companies')
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.tight_layout()
plt.show()
```

```python
# Now look in terms of sectors

data_sectors_tracking = pd.merge(data_w_sectors, df_tracking, on='Beauhurst URL')

data_sectors_tracking.groupby(['Sectors', 'Tracked status'], as_index=False).sum(numeric_only=True)

data_track_group = data_sectors_tracking.groupby(['Sectors', 'Tracked status'], as_index=False).agg({'Beauhurst URL':'nunique'})
```

```python
data_track_group.pivot(index='Sectors', columns='Tracked status', values='Beauhurst URL').to_clipboard()
```

```python
data_tracking_reasons = data_sectors_tracking.groupby(['Sectors', 'Tracked status'], as_index=False).sum(numeric_only=True)

data_tracking_reasons = data_tracking_reasons.loc[data_tracking_reasons['Tracked status'] == 'Tracked']
data_tracking_reasons.columns = data_tracking_reasons.columns.str.replace('Tracking reasons - ','')

data_tracking_reasons.drop(columns='Tracked status', inplace=True)

data_tracking_melt = data_tracking_reasons.melt(id_vars='Sectors', value_name='Count', var_name='Tracking reason').sort_values(by='Count', ascending=False)

data_tracking_melt.to_clipboard(index=False)
```

```python
fig, axs = plt.subplots(3,2, figsize=(10,10), sharey=True)

for ax, sector, c in zip(axs.flatten(), data_tracking_melt['Sectors'].unique(), colors):
    data_tmp = data_tracking_melt.loc[data_tracking_melt['Sectors'] == sector]
    sns.barplot(data=data_tmp, x='Count', y='Tracking reason', color=c, ax=ax, label=sector)

    ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))


# Create a single legend for all subplots
# First, collect all handles and labels
handles, labels = [], []
for ax in axs.flatten():
    h, l = ax.get_legend_handles_labels()
    handles.extend(h)
    labels.extend(l)

# Add the legend to the figure
fig.legend(handles, labels, loc='lower center', ncol=3, bbox_to_anchor=(0.5,1))

plt.tight_layout()

plt.show()
```

```python
# Select columns containing 'financial statement'
selected_columns = frame.copy().loc[:, frame.columns.str.contains('Financial statement 1 - ', case=False)]
selected_columns['Beauhurst URL'] = frame['Beauhurst URL']
```

```python
selected_columns.columns = selected_columns.columns.str.replace('Financial Statement 1 - ', '')
```

```python
selected_columns.columns
```

```python
selected_columns.isna().sum()
```

```python
selected_columns.dropna(subset='Date of accounts', inplace=True)
```

```python

selected_columns['Date of accounts'] = pd.to_datetime(selected_columns['Date of accounts'], format='%Y-%m-%d')

# Assuming your datetime column is already converted
selected_columns['financial_year'] = selected_columns['Date of accounts'].apply(
    lambda date: f"{date.year}-{date.year + 1}" if date.month >= 4 else f"{date.year - 1}-{date.year}"
)
```

```python
selected_columns['Accounts currency'].unique()
```

```python
selected_columns['Turnover'].sum() / 1e9
```

```python
data_financials = pd.merge(data_w_sectors, selected_columns, on='Beauhurst URL')
```

```python
data_turnover = data_financials.copy().dropna(subset='Turnover')

data_turnover['Turnover'] = data_turnover['Turnover'] / 1e6

data_turnover = data_turnover.loc[data_turnover['Date of accounts'].dt.year > 2020]

data_turnover_group = data_turnover.groupby(['Sectors', 'financial_year'], as_index=False).agg({'Beauhurst URL':'nunique', 'Turnover':'sum'}).sort_values(by=['financial_year', 'Sectors'])

plt.figure(figsize=(10,5))

sns.barplot(data=data_turnover_group, x='financial_year', y='Turnover', hue='Sectors')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.xlabel('Financial Year')
plt.ylabel('Turnover (£ millions)')

plt.tight_layout()
plt.show()
```

```python
data_turnover_group['average_turnover'] = data_turnover_group['Turnover'] / data_turnover_group['Beauhurst URL']

plt.figure(figsize=(10,5))

sns.barplot(data=data_turnover_group, x='financial_year', y='average_turnover', hue='Sectors')

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.xlabel('Financial Year')
plt.ylabel('Turnover (£ millions)')

plt.tight_layout()
plt.show()
```

```python
# Stage of company

data_sectors_stage = pd.merge(data_w_sectors, frame[['Beauhurst URL', 'Current Stage of Evolution', 'Tracked status']], on='Beauhurst URL')

data_stage_group = data_sectors_stage.groupby(['Sectors', 'Current Stage of Evolution', 'Tracked status'], as_index=False).nunique()

data_group = data_stage_group.groupby(['Sectors', 'Current Stage of Evolution'], as_index=False).sum(numeric_only=True)

plt.figure(figsize=(12,5))

sns.barplot(data=data_group, x='Sectors', y='Beauhurst URL', hue='Current Stage of Evolution')

plt.ylabel('Number of companies')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
```

```python
data_group['Percentage'] = data_group.groupby('Sectors')['Beauhurst URL'].transform(lambda x: x / x.sum() * 100)

plt.figure(figsize=(12,5))

sns.barplot(data=data_group, x='Sectors', y='Percentage', hue='Current Stage of Evolution')

plt.ylabel('Percentage (%)')
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))

plt.legend(bbox_to_anchor=(1,1))

plt.tight_layout()
```

```python
data_fundraising = pd.read_csv('beauhurst_fundraising_export_20250910-163916.csv')
```

```python
print(len(data_fundraising))

print(len(data_fundraising['Beauhurst company URL'].drop_duplicates()))
```

```python
data_fundraising_sectors = pd.merge(data_fundraising, data_w_sectors, left_on='Beauhurst company URL', right_on='Beauhurst URL')
```

```python
# fundraisings over time

data_fundraising_sectors['deal_year'] = pd.to_datetime(data_fundraising_sectors['Deal date'], format='%Y-%m-%d').dt.year

data_fundraising_group = data_fundraising_sectors.groupby(['deal_year', 'Sectors', 'Stage of evolution at deal date'], as_index=False).agg({'Beauhurst deal URL':'count', 'Amount raised':'sum'})
data_fundraising_group['Amount raised (millions)'] = data_fundraising_group['Amount raised'] / 1e6
```

```python
data_group = data_fundraising_group.groupby(['deal_year', 'Sectors'], as_index=False).sum(numeric_only=True)

data_group = data_group.loc[data_group['deal_year'] > 2015]
```

```python
fig, axs = plt.subplots(3,2, figsize=(10,10), sharex=True)

for ax, sector, c in zip(axs.flatten(), data_group['Sectors'].unique(), colors):
    data_tmp = data_group.loc[data_group['Sectors'] == sector]
    sns.barplot(data=data_tmp, y='Amount raised (millions)', x='deal_year', color=c, ax=ax, label=sector, width=0.6)

    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    ax.set_xlabel("Deal year")


# Create a single legend for all subplots
# First, collect all handles and labels
handles, labels = [], []
for ax in axs.flatten():
    h, l = ax.get_legend_handles_labels()
    handles.extend(h)
    labels.extend(l)

# Add the legend to the figure
fig.legend(handles, labels, loc='lower center', ncol=3, bbox_to_anchor=(0.5,1))

plt.tight_layout()

plt.show()
```
